```java
import java.util.*;

// Queue Calculator
Queue<String> queueOps = new LinkedList<>();
queueOps.offer("10");
queueOps.offer("+");
queueOps.offer("5");
queueOps.offer("*");
queueOps.offer("2");

System.out.println("Queue Operations: " + queueOps);

double queueResult = 0;
boolean firstQueue = true;

while (!queueOps.isEmpty()) {
    String token = queueOps.poll();
    if (token.matches("-?\\d+(\\.\\d+)?")) {
        if (firstQueue) {
            queueResult = Double.parseDouble(token);
            firstQueue = false;
        } else {
            System.out.println("Error in Queue!");
        }
    } else {
        if (queueOps.isEmpty()) break;
        String nextNum = queueOps.poll();
        double num = Double.parseDouble(nextNum);
        switch (token) {
            case "+": queueResult += num; break;
            case "-": queueResult -= num; break;
            case "*": queueResult *= num; break;
            case "/": queueResult /= num; break;
            default: System.out.println("Unknown operation in Queue!");
        }
        System.out.println("Current Queue Result: " + queueResult);
    }
}

System.out.println("Final Queue Result: " + queueResult);

// Stack Calculator
Stack<String> stackOps = new Stack<>();
stackOps.push("2");
stackOps.push("*");
stackOps.push("5");
stackOps.push("+");
stackOps.push("10");

System.out.println("\nStack Operations: " + stackOps);

Stack<String> tempStack = new Stack<>();
while (!stackOps.isEmpty()) {
    tempStack.push(stackOps.pop());
}

double stackResult = 0;
boolean firstStack = true;

while (!tempStack.isEmpty()) {
    String token = tempStack.pop();
    if (token.matches("-?\\d+(\\.\\d+)?")) {
        if (firstStack) {
            stackResult = Double.parseDouble(token);
            firstStack = false;
        } else {
            System.out.println("Error in Stack!");
        }
    } else {
        if (tempStack.isEmpty()) break;
        String nextNum = tempStack.pop();
        double num = Double.parseDouble(nextNum);
        switch (token) {
            case "+": stackResult += num; break;
            case "-": stackResult -= num; break;
            case "*": stackResult *= num; break;
            case "/": stackResult /= num; break;
            default: System.out.println("Unknown operation in Stack!");
        }
        System.out.println("Current Stack Result: " + stackResult);
    }
}

System.out.println("Final Stack Result: " + stackResult);

// Postfix Calculator
Stack<Double> postfixStack = new Stack<>();
String[] postfix = {"10", "2", "/", "3", "*"};

for (int i = 0; i < postfix.length; i++) {
    String token = postfix[i];
    if (token.matches("-?\\d+(\\.\\d+)?")) {
        postfixStack.push(Double.valueOf(token));
    } else {
        if (postfixStack.size() < 2) continue;
        double a = postfixStack.pop();
        double b = postfixStack.pop();
        switch (token) {
            case "^": postfixStack.push(Math.pow(b, a)); break;
            case "*": postfixStack.push(b * a); break;
            case "/": postfixStack.push(b / a); break;
            case "+": postfixStack.push(b + a); break;
            case "-": postfixStack.push(b - a); break;
            default:
                postfixStack.push(b);
                postfixStack.push(a);
        }
    }
}

System.out.println("\nPostfix Result: " + postfixStack.peek());
```
